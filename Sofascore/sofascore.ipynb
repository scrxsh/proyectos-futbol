```python
shots = sofascore.get_match_shotmap('https://www.sofascore.com/es-la/football/match/real-madrid-real-sociedad/zgbsEgb#id:12437506')
shots
```

```python
fig, ax = plt.subplots(figsize=(7, 6.5))
fig.set_facecolor('#47832d')
ax.patch.set_facecolor('#332a49') 

pitch = Pitch(pitch_type='opta', pitch_color='grass', stripe=True, line_color='white')
pitch.draw(ax=ax)  # Dibuja primero el campo

# Ahora los scatter, así quedan por encima del campo
local = shots[(shots['isHome'] == True) & (shots['shotType'].isin(['miss', 'block']))].copy()
local['x'] = 100 - local['x']
plt.scatter(local['x'], local['y'], c='#fff', s=110, alpha=.7, linewidths=0.5, edgecolors='#000')

local = shots[(shots['isHome'] == True) & (shots['shotType'] == 'save')].copy()
local['x'] = 100 - local['x']
plt.scatter(local['x'], local['y'], c='#fff', s=130, alpha=.8, linewidths=1, edgecolors='#000')

local = shots[(shots['isHome'] == True) & (shots['shotType'] == 'goal')].copy()
local['x'] = 100 - local['x']
plt.scatter(local['x'], local['y'], c='#fff', s=160, linewidths=1.5, edgecolors='#000')

visitante = shots[(shots['isHome'] != True) & (shots['shotType'].isin(['miss', 'block']))].copy()
plt.scatter(visitante['x'], visitante['y'], c="#3601C7", s=110, alpha=.7, edgecolors='#fff', linewidths=0.5)

visitante = shots[(shots['isHome'] != True) & (shots['shotType'] == 'save')].copy()
plt.scatter(visitante['x'], visitante['y'], c="#3601C7", s=130, alpha=.8, edgecolors='#fff', linewidths=1)

visitante = shots[(shots['isHome'] != True) & (shots['shotType'] == 'goal')].copy()
plt.scatter(visitante['x'], visitante['y'], c="#3601C7", s=160, edgecolors='#fff', linewidths=1.5)

plt.gca().invert_xaxis()
plt.title('Mapa de disparos del Real Madrid - Real Sociedad', fontsize=15, c='#fff')
```
