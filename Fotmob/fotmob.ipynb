```python
import LanusStats as ls
import matplotlib.pyplot as plt
from mplsoccer.pitch import VerticalPitch, Pitch
fotmob = ls.FotMob()
```

```python
data = fotmob.get_match_shotmap(4507097)
data
```

```python
fig, ax = plt.subplots(figsize=(8, 6.5))
fig.set_facecolor('#332a49')
ax.patch.set_facecolor('#332a49') 

pitch = Pitch(pitch_type='uefa', line_color='#cfcfcf', line_zorder=3, pitch_color='#332a49')

pitch.draw(ax=ax)
plt.scatter(data['x'], data['y'], c='#fff', s=100, alpha=.5)
plt.gca().invert_yaxis()

plt.title('Mapa de disparos del Barcelona - Real Madrid', fontsize=15, c='#fff')
```

```python
final_uel = fotmob.get_match_shotmap(4737728)
final_uel.to_csv('csv/final_uel.csv', index=False)
final_uel
```

```python
fig, ax = plt.subplots(figsize=(8, 6.5))
fig.set_facecolor('#47832d')
ax.patch.set_facecolor('#332a49') 

pitch = Pitch(pitch_type='uefa', line_zorder=0,  pitch_color='grass', stripe=True, line_color='white')


# Disparos del equipo local (8586) en la mitad izquierda
local = final_uel[final_uel['teamId'] == 8586]
local['x'] = 105 - local['x']  # Refleja los disparos a la izquierda
plt.scatter(local['x'], local['y'], c='#fff', s=100)

# Disparos del equipo visitante (10260) en la mitad derecha
visitante = final_uel[final_uel['teamId'] != 8586]
plt.scatter(visitante['x'], visitante['y'], c='#C70101', s=100)

pitch.draw(ax=ax)
plt.gca().invert_yaxis()


plt.title('Mapa de disparos del Tottenham - Man. Utd', fontsize=18, c='#fff')
```

```python
fig, ax = plt.subplots(figsize=(9.5, 9.5))
fig.set_facecolor('#47832d')
ax.patch.set_facecolor('#332a49')

pitch = VerticalPitch(pitch_type='uefa', line_zorder=0,  pitch_color='grass', stripe=True, line_color='white')


# Disparos del equipo local (8586) en la mitad superior
local['y'] = 68 - local['y']  # Refleja los disparos hacia arriba
plt.scatter(local['y'], local['x'], c='#fff', s=100)

# Disparos del equipo visitante (10260) en la mitad inferior
plt.scatter(visitante['y'], visitante['x'], c='#C70101', s=100)

pitch.draw(ax=ax)

plt.title('Mapa de disparos del Tottenham - Man. Utd', fontsize=18, c='#fff')
```
